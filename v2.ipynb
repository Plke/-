```python
import requests
from bs4 import BeautifulSoup
import re
# 发送HTTP请求获取网页内容

book_name="战争世界马旒斯"
url_root = "http://www.xbiquzw.com/"
search_response = requests.get(url_root+"modules/article/search.php?searchkey="+book_name)
# 使用Beautiful Soup解析HTML内容
search_soup = BeautifulSoup(search_response.text, "html.parser")
#获取"#wrapper > table > tbody" tbody
tbody = search_soup.select_one("#wrapper>table")
```

```python
#遍历tbody下的tr，从第二个开始
for tr in tbody.find_all('tr'):
    #获取第一个td
    td = tr.find('td')
    if(td==None):
        continue
    #获取a标签
    a = td.find('a')
    #获取内容
    book_title = a.get_text()
    if(book_title==book_name):
        #获取href
        href = a.get('href')
        #跳转
        book_url = url_root+href
        book_response = requests.get(book_url)
        book_soup = BeautifulSoup(book_response.text, "html.parser",from_encoding="utf-8")
        #list > dl
        dl = book_soup.select_one("#list>dl")
        #遍历dl下的dd
        with open("book/"+book_name+".txt", "a", encoding="utf-8") as f:
            for dd in dl.find_all('dd'):
                #获取a标签
                a = dd.find('a')
                #获取内容
                href = a.get('href')

                chapter_title = a.get('title').encode("latin1").decode("utf-8")

                print(chapter_title)

                chapter_url=book_url+href
                chapter_response = requests.get(chapter_url)
                chapter_soup = BeautifulSoup(chapter_response.text, "html.parser",from_encoding="utf-8")
                content = chapter_soup.select_one("#content")
                #获取内容,包括html
                #获取div中的内容
                
                content_text = content.encode("latin1").decode("utf-8", errors="ignore")
                pattern = re.compile(r'<!--go-->\s*(.*?)\s*<!--over-->', re.S)
                content_text = pattern.search(content_text).group(1).strip()
                #写入文件
                
                f.write("\n\n"+chapter_title+'\n\n')
                f.write(content_text.replace("<br/>", "\n"))

```
